<h2>First Try of Feature Engineering</h2>

- Focus will be on testing the coding and engineering the climate risks (wildfires, heatwave, storms, coldwave, and flood)

```python
import pandas as pd
```

```python
raw_data = pd.read_csv('corn_climate_risk_futures_daily_master.csv')
raw_data.head(5)
```

```python
raw_data.shape
```

```python
climate_cols = [c for c in raw_data.columns if c.startswith('climate_risk_')]
futures_cols = [c for c in raw_data.columns if c.startswith('futures_')]

climate_cols
```

```python
# Climate Risks for Wildfires, and Storms: 

raw_data['climate_risk_cnt_locations_wildfires_risk_medium'] = raw_data[['climate_risk_cnt_locations_heat_stress_risk_medium', 'climate_risk_cnt_locations_drought_risk_medium']].min(axis=1)
raw_data['climate_risk_cnt_locations_wildfires_risk_high'] = raw_data[['climate_risk_cnt_locations_heat_stress_risk_high', 'climate_risk_cnt_locations_drought_risk_high']].min(axis=1)

raw_data['climate_risk_cnt_locations_storms_risk_medium'] = raw_data[['climate_risk_cnt_locations_heat_stress_risk_medium', 'climate_risk_cnt_locations_excess_precip_risk_medium']].min(axis=1)
raw_data['climate_risk_cnt_locations_storms_risk_high'] = raw_data[['climate_risk_cnt_locations_heat_stress_risk_high', 'climate_risk_cnt_locations_excess_precip_risk_high']].min(axis=1)
```

```python
raw_data_2 = raw_data.copy()
raw_data_2.head()
```

```python
#Climate Risk for Heat wave, Cold wave, and Flood
for i in range(1, 5):
    raw_data_2[f'medium_heatstress_lag_{i}'] = raw_data_2['climate_risk_cnt_locations_heat_stress_risk_medium'].shift(i)
    raw_data_2[f'medium_coldstress_lag_{i}'] = raw_data_2['climate_risk_cnt_locations_unseasonably_cold_risk_medium'].shift(i)
    raw_data_2[f'medium_precip_lag_{i}'] = raw_data_2['climate_risk_cnt_locations_excess_precip_risk_medium'].shift(i)
    
    raw_data_2[f'medium_heatstress_lag_{i}'] = raw_data_2[f'medium_heatstress_lag_{i}'].fillna(0)
    raw_data_2[f'medium_coldstress_lag_{i}'] = raw_data_2[f'medium_coldstress_lag_{i}'].fillna(0)
    raw_data_2[f'medium_precip_lag_{i}'] = raw_data_2[f'medium_precip_lag_{i}'].fillna(0)

for j in range(1, 3): 
    raw_data_2[f'high_heatstress_lag_{j}'] = raw_data_2['climate_risk_cnt_locations_heat_stress_risk_high'].shift(j)
    raw_data_2[f'high_coldstress_lag_{j}'] = raw_data_2['climate_risk_cnt_locations_unseasonably_cold_risk_high'].shift(j)
    raw_data_2[f'high_precip_lag_{j}'] = raw_data_2['climate_risk_cnt_locations_excess_precip_risk_high'].shift(j)

    raw_data_2[f'high_heatstress_lag_{j}'] = raw_data_2[f'high_heatstress_lag_{j}'].fillna(0)
    raw_data_2[f'high_coldstress_lag_{j}'] = raw_data_2[f'high_coldstress_lag_{j}'].fillna(0)
    raw_data_2[f'high_precip_lag_{j}'] = raw_data_2[f'high_precip_lag_{j}'].fillna(0)
```

```python
raw_data_2['total_location_by_region'] = raw_data_2['climate_risk_cnt_locations_heat_stress_risk_low'] + raw_data_2['climate_risk_cnt_locations_heat_stress_risk_medium'] \
                                        + raw_data_2['climate_risk_cnt_locations_heat_stress_risk_high']
```

```python
raw_data_2['medium_heatstress_4days_average'] = raw_data_2[[f'medium_heatstress_lag_{i}' for i in range(1, 5)]].mean(axis=1)
raw_data_2['medium_heatstress_2days_average'] = raw_data_2[[f'medium_heatstress_lag_{i}' for i in range(1, 3)]].mean(axis=1)
raw_data_2['high_heatstress_2days_average'] = raw_data_2[[f'high_heatstress_lag_{i}' for i in range(1, 3)]].mean(axis=1)

raw_data_2['medium_coldstress_4days_average'] = raw_data_2[[f'medium_coldstress_lag_{i}' for i in range(1, 5)]].mean(axis=1)
raw_data_2['medium_coldstress_2days_average'] = raw_data_2[[f'medium_coldstress_lag_{i}' for i in range(1, 3)]].mean(axis=1)
raw_data_2['high_coldstress_2days_average'] = raw_data_2[[f'high_coldstress_lag_{i}' for i in range(1, 3)]].mean(axis=1)

raw_data_2['medium_precip_4days_average'] = raw_data_2[[f'medium_precip_lag_{i}' for i in range(1, 5)]].mean(axis=1)
raw_data_2['medium_precip_2days_average'] = raw_data_2[[f'medium_precip_lag_{i}' for i in range(1, 3)]].mean(axis=1)
raw_data_2['high_precip_2days_average'] = raw_data_2[[f'high_precip_lag_{i}' for i in range(1, 3)]].mean(axis=1)

raw_data_2['climate_risk_cnt_locations_heatwave_risk_high'] = (raw_data_2['medium_heatstress_4days_average'] + raw_data_2['high_heatstress_2days_average']) / 2
raw_data_2['climate_risk_cnt_locations_heatwave_risk_medium'] = (raw_data_2['medium_heatstress_2days_average'] + raw_data_2['high_heatstress_lag_1']) / 2

raw_data_2['climate_risk_cnt_locations_coldwave_risk_high'] = (raw_data_2['medium_coldstress_4days_average'] + raw_data_2['high_coldstress_2days_average']) / 2
raw_data_2['climate_risk_cnt_locations_coldwave_risk_medium'] = (raw_data_2['medium_coldstress_2days_average'] + raw_data_2['high_coldstress_lag_1']) / 2

raw_data_2['climate_risk_cnt_locations_flood_risk_high'] = (raw_data_2['medium_precip_4days_average'] + raw_data_2['high_precip_2days_average']) / 2
raw_data_2['climate_risk_cnt_locations_flood_risk_medium'] = (raw_data_2['medium_precip_2days_average'] + raw_data_2['high_precip_lag_1']) / 2

raw_data_2
```
